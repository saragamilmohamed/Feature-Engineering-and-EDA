```python
df['datetime']=pd.to_datetime(df['datetime'])
```

```python
df['Year']=df['datetime'].dt.year
```

```python
df['Month']=df['datetime'].dt.month
```

```python
df['MonthName']=df['datetime'].dt.month_name()
```

```python
df['Day']=df['datetime'].dt.day
```

```python
df['DayName']=df['datetime'].dt.day_name()
```

```python
df['week']=df['datetime'].dt.isocalendar().week
```

```python
df['WeekDay']=df['datetime'].dt.weekday
```

```python
df['Hours']=df['datetime'].dt.hour
```

```python
df['Minutes']=df['datetime'].dt.minute
```

```python
df['Second']=df['datetime'].dt.second
```

```python
sns.countplot(x=df['Day'])
```

```python
sns.countplot(x=df['WeekDay'])
```

```python
def period_day(x):
    if x in range(0, 13):
        return 'morning'
    elif x in range(13, 19):
        return 'afternoon'
    else:
        return 'evening'
```

```python
df['Period ']=df['Hours'].apply(period_day)
```

```python
sns.countplot(x=df['Period '])
```

```python
sns.barplot(x=df['Period '],y=df['Hours'])
```

```python
def Seasons(x):
    if x in [12, 1, 2]:
        return 'Winter'
    elif x in [3, 4, 5]:
        return 'Spring'
    elif x in [6, 7, 8]:
        return 'Summer'
    elif x in [9, 10, 11]:
        return 'Autumn'
```

```python
df['Seasons']=df['Month'].apply(Seasons)
```

```python
sns.countplot(x=df['Seasons'])
```

```python
sns.barplot(x=df['Seasons'],y=df['Month'])
```

```python
data=pd.read_csv('chicago.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python

def street(street):
    return street.split('&')[0]
```

```python
data['Street']=data['Start Station'].apply(street)
data['Street']
```

```python
data['Street'].value_counts()
```

```python
plt.figure(figsize=(18,5),dpi=100)
sns.countplot(x=data['Street'])
plt.xticks(rotation=90)
plt.show
```

```python

def neighborhood(neighborhood):
    return neighborhood.split('&')[-1]
```

```python
data['Neighborhood']=data['Start Station'].apply( neighborhood)
```

```python
data['Neighborhood']
```

```python
plt.figure(figsize=(18,5),dpi=100)
sns.countplot(x=data['Neighborhood'])
plt.xticks(rotation=90)
plt.show
```
